```python
display(planilha)
planilha['Salário']>2000 # retorna booleano se o salario é ou nao maior que 2000
```

```python
planilha.loc[planilha['Salário'] > 2000] #retorna apenas o salarios que atendem a condição estipulada
```

```python
planilha.loc[planilha['Data_Nasc'] > '2001-01-01'] #comparação com data
```

```python
# and == &

planilha.loc[(planilha['Cidade'] == 'Sorocaba') & (planilha['Salário'] > 2000)] 
```

```python
# or == |

planilha.loc[(planilha['Cidade'] == 'Sorocaba') | (planilha['Cidade'] == 'Votorantim')]
```

```python
# FUNÇÕES DE AGREGAÇÃO

print(planilha['Salário'].sum()) # soma dos salários
print(planilha['Salário'].max()) # maior salário
print(planilha['Salário'].min()) # menor salário
print(planilha['Salário'].mean()) # méida do salários
```

```python
#Agrupar

print(planilha.groupby('Cidade')['Salário'].mean())
```

```python
# criar coluna

planilha['Salário+10%'] = planilha ['Salário'] * 1.1

display(planilha)
```

```python
# criando um coluna para exclui-la depois

planilha ['imposto'] = 8

display(planilha)
```

```python
# Excluindo a coluna

planilha = planilha.drop(columns=['imposto'])

display(planilha)
```

```python
# exportar o arquivo criado

planilha.to_excel('Nova planilha.xlsx')
```

```python
planilha ['Cidade/UF'] = planilha['Cidade'] + '/SP'

display(planilha)
```

```python
planilha.to_excel('Nova planilha.xlsx')
```
